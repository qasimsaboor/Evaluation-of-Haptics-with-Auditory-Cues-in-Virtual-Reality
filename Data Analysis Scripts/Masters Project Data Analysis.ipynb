```python
#Computation for spatial presence
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon
from scipy.stats import zscore

# Load the haptics data
haptics_data = pd.read_csv('D:\Research Project\Haptics.csv')

# Load the without haptics data
without_haptics_data = pd.read_csv('D:\Research Project\Without_Haptics.csv')

# Assuming the column names for spatial presence ratings are the same as the column_names list
spatial_presence_column_names = [
    'I felt like I was actually there in the environment of the presentation',
    'It seemed as though I actually took part in the action of the presentation',
    'It was as though my true location had shifted into the environment in the presentation',
    'I felt as though I was physically present in the environment of the presentation',
    'I experienced the environment in the presentation as though I had stepped into a different place',
    'I was convinced that things were actually happening around me',
    'I had the feeling that I was in the middle of the action rather than merely observing',
    'I felt like the objects in the presentation surrounded me',
    'I experienced both the confined and open spaces in the presentation as though I was really there',
    'I was convinced that the objects in the presentation were located on the various sides of my body'
]

# Convert ratings data to numeric format
spatial_presence_haptics = haptics_data[spatial_presence_column_names].apply(pd.to_numeric, errors='coerce')
spatial_presence_without_haptics = without_haptics_data[spatial_presence_column_names].apply(pd.to_numeric, errors='coerce')

# Calculate the overall rating for each participant by averaging across all columns
haptics_ratings = spatial_presence_haptics.mean(axis=1)
without_haptics_ratings = spatial_presence_without_haptics.mean(axis=1)

# Perform the Wilcoxon signed-rank test on the overall ratings
_, p_value = wilcoxon(haptics_ratings.dropna(), without_haptics_ratings.dropna())

# Create a bar plot to compare mean spatial presence ratings
labels = ['Haptics', 'Without Haptics']
means = [haptics_ratings.mean(), without_haptics_ratings.mean()]
std_devs = [haptics_ratings.std(), without_haptics_ratings.std()]
print(means)
print(std_devs)
colors = ['blue', 'green']  # Add custom colors here

plt.bar(labels, means, color=colors,yerr=std_devs, capsize=10)
plt.ylabel('Mean Spatial Presence Rating')
plt.title('Comparison of Spatial Presence Ratings\nHaptics vs. Without Haptics')
plt.ylim(1, 7)  # Set the y-axis limits to match the Likert scale range
plt.show()

# Print the hypothesis being tested and the result
print(p_value)
hypothesis = "Null: With haptic feedback, users will report higher levels of spatial presence compared to without haptic feedback."
print("Hypothesis: {}".format(hypothesis))
if p_value < 0.05:
    print("Result: Hypothesis is significant.")
else:
    print("Result: Hypothesis is not significant.")
```

```python
#To compute user experince qualities, pragmatic, hedonic and overall attractiveness
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon
from scipy.stats import zscore

# Load the haptics data
haptics_data = pd.read_csv('D:\Research Project\Haptics.csv')

# Load the without haptics data
without_haptics_data = pd.read_csv('D:\Research Project\Without_Haptics.csv')

# Define a function to convert ratings from -3 to +3 scale to 1 to 7 scale
def convert_to_1_to_7_scale(rating):
    return (rating + 3) * (7 / 6)


# Assuming the column names for the AttrakDiff ratings are the same as the attrakdiff_column_names list
attrakdiff_column_names = [
        'Complicated/Simple',
        'Impractical/Practical',
        'Predictable/Unpredictable',
        'Confusing/Clear',
        'Dull/Captivating',
        'Creative/Unimaginative',
        'Appealing/Repelling',
        'Premium/Cheap',
        'Ugly/Attractive',
        'Bad/Good'
]


# Group the columns into three categories
pragmatic_columns = attrakdiff_column_names[:4]
hedonic_columns = attrakdiff_column_names[4:8]
overall_columns = attrakdiff_column_names[8:]

# Convert ratings data to numeric format for each category
attrakdiff_haptics_pragmatic = haptics_data[pragmatic_columns].apply(pd.to_numeric, errors='coerce')
attrakdiff_haptics_hedonic = haptics_data[hedonic_columns].apply(pd.to_numeric, errors='coerce')
attrakdiff_haptics_overall = haptics_data[overall_columns].apply(pd.to_numeric, errors='coerce')

attrakdiff_without_haptics_pragmatic = without_haptics_data[pragmatic_columns].apply(pd.to_numeric, errors='coerce')
attrakdiff_without_haptics_hedonic = without_haptics_data[hedonic_columns].apply(pd.to_numeric, errors='coerce')
attrakdiff_without_haptics_overall = without_haptics_data[overall_columns].apply(pd.to_numeric, errors='coerce')

# Calculate the mean ratings for each category for both groups
haptics_ratings_pragmatic = attrakdiff_haptics_pragmatic.mean(axis=1)
haptics_ratings_hedonic = attrakdiff_haptics_hedonic.mean(axis=1)
haptics_ratings_overall = attrakdiff_haptics_overall.mean(axis=1)

print(haptics_ratings_pragmatic.mean())
print(haptics_ratings_hedonic.mean())
print(haptics_ratings_overall.mean())

without_haptics_ratings_pragmatic = attrakdiff_without_haptics_pragmatic.mean(axis=1)
without_haptics_ratings_hedonic = attrakdiff_without_haptics_hedonic.mean(axis=1)
without_haptics_ratings_overall = attrakdiff_without_haptics_overall.mean(axis=1)

print(without_haptics_ratings_pragmatic.mean())
print(without_haptics_ratings_hedonic.mean())
print(without_haptics_ratings_overall.mean())

# Perform Wilcoxon signed-rank tests on the overall ratings for each category
_, p_value_pragmatic = wilcoxon(haptics_ratings_pragmatic.dropna(), without_haptics_ratings_pragmatic.dropna())
_, p_value_hedonic = wilcoxon(haptics_ratings_hedonic.dropna(), without_haptics_ratings_hedonic.dropna())
_, p_value_overall = wilcoxon(haptics_ratings_overall.dropna(), without_haptics_ratings_overall.dropna())


# Calculate the standard deviations for each category for both groups
haptics_std_devs = [haptics_ratings_pragmatic.std(), haptics_ratings_hedonic.std(), haptics_ratings_overall.std()]
without_haptics_std_devs = [without_haptics_ratings_pragmatic.std(), without_haptics_ratings_hedonic.std(), without_haptics_ratings_overall.std()]
print(haptics_std_devs)
print(without_haptics_std_devs)

# Create a bar plot to compare mean ratings for each category for both groups
labels = ['Pragmatic Quality', 'Hedonic Quality', 'Overall Attractiveness']
bar_width = 0.2
index_haptics = range(len(labels))
index_without_haptics = [i + bar_width for i in index_haptics]

fig, ax = plt.subplots()

ax.bar(index_haptics, [haptics_ratings_pragmatic.mean(), haptics_ratings_hedonic.mean(), haptics_ratings_overall.mean()],
       bar_width, yerr=haptics_std_devs, capsize=10, label='Haptics', color='blue')
ax.bar(index_without_haptics, [without_haptics_ratings_pragmatic.mean(), without_haptics_ratings_hedonic.mean(), without_haptics_ratings_overall.mean()],
       bar_width, yerr=without_haptics_std_devs, capsize=10, label='Without Haptics', color='orange')

ax.set_xlabel('Category')
ax.set_ylabel('Mean AttrakDiff Rating')
ax.set_title('Comparison of User Experience Ratings\nHaptics vs. Without Haptics')
ax.set_ylim(1, 7)  # Set the y-axis limits to match the AttrakDiff scale range
ax.legend()

# Set the x-axis labels to categories and position them at the center of the grouped bars
ax.set_xticks([i + bar_width / 2 for i in index_haptics])
ax.set_xticklabels(labels)

plt.show()

print(p_value_pragmatic)
print(p_value_hedonic)
print(p_value_overall)
```

```python
# all 4 time tasks computation using Wilcoxon signed-rank test
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import wilcoxon

def read_csv_file(file_path):
    return pd.read_csv(file_path)

def compute_average_time(df, task_columns):
    # Assuming the 'Participant ID' column exists in the DataFrame
    user_column = 'Participant ID'

    # Compute the average time for each task
    avg_times = df.groupby(user_column)[task_columns].mean()

    return avg_times

def perform_wilcoxon_signed_rank_test(haptics_avg, without_haptics_avg):
    p_values = []
    for task in haptics_avg.columns:
        _, p_value = wilcoxon(haptics_avg[task], without_haptics_avg[task])
        p_values.append(p_value)
    return p_values

def plot_average_times(haptics_avg, without_haptics_avg):
    tasks = ['Task 1', 'Task 2', 'Task 3', 'Task 4']
    bar_width = 0.2
    index = np.arange(len(tasks))

    plt.bar(index, haptics_avg, width=bar_width, label='Haptics')
    plt.bar(index + bar_width, without_haptics_avg, width=bar_width, label='Without Haptics')

    plt.xlabel('Tasks')
    plt.ylabel('Average Time (seconds)')
    plt.title('Average Time Taken for All Tasks')
    plt.xticks(index + bar_width / 2, tasks)
    plt.legend()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    file1 = "D:\Research Project\Haptics.csv"
    file2 = "D:\Research Project\Without_Haptics.csv"

    # Assuming the column names for Task times are 'Task 1 time', 'Task 2 time', 'Task 3 time', 'Task 4 time'
    task_columns = ['Task 1 time', 'Task 2 time', 'Task 3 time', 'Task 4 time']

    # Read the CSV files
    df1 = read_csv_file(file1)
    df2 = read_csv_file(file2)

    # Compute average time for all tasks with haptics and without haptics
    haptics_avg = compute_average_time(df1, task_columns)
    without_haptics_avg = compute_average_time(df2, task_columns)

    # Perform the Wilcoxon signed-rank test for each task
    p_values = perform_wilcoxon_signed_rank_test(haptics_avg, without_haptics_avg)
    print("P-values for each task:", p_values)

    # Plot the results
    plot_average_times(haptics_avg.mean(), without_haptics_avg.mean())
```

```python
#Combined time computation for all 4 tasks using Wilcoxon signed-rank test
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon

def read_csv_file(file_path):
    return pd.read_csv(file_path)

def compute_overall_average(df, task_columns):
    # Assuming the 'Participant ID' column exists in the DataFrame
    user_column = 'Participant ID'

    # Compute the sum of time for all tasks
    df['Total Time'] = df[task_columns].sum(axis=1)

    # Compute the overall average for each user
    overall_avg = df.groupby(user_column)['Total Time'].mean()

    return overall_avg

def perform_wilcoxon_signed_rank_test(haptics_avg, without_haptics_avg):
    _, p_value = wilcoxon(haptics_avg, without_haptics_avg)
    return p_value

def plot_overall_average(haptics_avg, without_haptics_avg):
    conditions = ['Haptics', 'Without Haptics']
    avg_times = [haptics_avg.mean(), without_haptics_avg.mean()]

    # Set the width for the bars to be displayed in parallel
    bar_width = 0.5

    # Set the positions of the bars on the x-axis
    index = [0, 1]

    # Create a bar graph with two sets of bars using different colors
    plt.bar(index, avg_times, width=bar_width, color=['blue', 'orange'])

    plt.xlabel('Conditions')
    plt.ylabel('Overall Average Time (seconds)')
    plt.title('Overall Average Time Taken for All Tasks')
    plt.xticks(index, conditions)
    plt.show()

if __name__ == "__main__":
    file1 = "D:\Research Project\Haptics.csv"
    file2 = "D:\Research Project\Without_Haptics.csv"

    # Assuming the column names for Task times are 'Task 1 time', 'Task 2 time', 'Task 3 time', 'Task 4 time'
    task_columns = ['Task 1 time', 'Task 2 time', 'Task 3 time', 'Task 4 time']

    # Read the CSV files
    df1 = read_csv_file(file1)
    df2 = read_csv_file(file2)

    # Compute overall average time for all tasks with haptics and without haptics
    haptics_avg = compute_overall_average(df1, task_columns)
    without_haptics_avg = compute_overall_average(df2, task_columns)

    # Perform the Wilcoxon signed-rank test on the overall average time for all tasks
    p_value = perform_wilcoxon_signed_rank_test(haptics_avg, without_haptics_avg)
    print("P-value:", p_value)

    # Plot the results
    plot_overall_average(haptics_avg, without_haptics_avg)
```
